```python
import pandas as pd
```

Import CSV

```python
df=pd.read_csv("../data/DCA_Legally_Operating_Businesses_08092021.csv")
```

```python
df
```

Change License Expiration Date to Date Time

```python
df["License Expiration Date"].apply(pd.to_datetime)
```

```python
df["License Expiration Date"] = df["License Expiration Date"].apply(pd.to_datetime)
```

```python
df.sort_values(by= "License Expiration Date")
```

Make the Date Time change permanent. 

```python
df = df.sort_values(by= "License Expiration Date", ascending=False)
```

Remove Duplicates within Business Name

```python
df.drop_duplicates(subset=["Business Name"])
```

```python
df=df.drop_duplicates(subset=["Business Name"])
```

```python
df.columns
```

Remove dates from before 2018

```python
df_filtered=df.loc[(df["License Expiration Date"]) >= "2018-01-01"]
```

Find the highest counts of Inactive licenses by zip code

```python
df_filtered[["License Status", "Address ZIP", "Address Borough"]]
```

```python
df_filtered[["License Status", "Address ZIP", "Address Borough", "Business Name"]].groupby(["Address Borough", "Address ZIP", "License Status"]).count()
```

```python
active_inactive_count = df_filtered[["License Status", "Address ZIP", "Address Borough", "Business Name"]].groupby(["Address Borough", "Address ZIP", "License Status"]).count().reset_index()
```

```python
active_inactive_count
```

```python
active_inactive_count
```

```python
active_inactive_count[active_inactive_count["License Status"].eq("Inactive")].sort_values(by="Business Name", ascending=False).head(10)
```

Determine turnover— Find addresses with unique business owners

```python
df_filtered.groupby(["Address Building", "Address Street Name", "Address ZIP"]).count().sort_values(by = "DCA License Number", ascending = False)
```

Zip Code 11216 has both one of the highest counts of inactive licenses in our date range, and an address with 4 different business owners. 

We will focus on that zip code to begin with. 

Bring up a new data frame showing just 11216.

```python
df_filtered.columns
```

```python
df_filtered[df_filtered["Address ZIP"].eq(11216)]
```

```python
df_11216 = df_filtered[df_filtered["Address ZIP"].eq(11216)]
```

```python
df_11216
```

Export to CSV

```python
df_11216.to_csv('../output/DCA_License_11216.csv', index = False)
```
